```python
from folium import plugins
```

```python
minimap = plugins.MiniMap()
m.add_child(minimap)
m
```

```python
def gps_to_feature(gps):
    return {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [gps['gps'][1], gps['gps'][0]]
        },
        "properties": {
            "time": gps['timestamp'],
            "style": {"color": ""},
            "icon": "circle",
            "iconstyle": {
                'fillColor': '#0000FF',
                'fillOpacity': 0.8,
                'stroke': 'true',
                'radius': 5
            }
        }
    }
```

```python
anim = plugins.TimestampedGeoJson(
    data=open('tmb_geojson.json'),
    period="PT1M",
    date_options="YYYY-MM-DD HH:mm:ss"
)
m.add_child(anim)
```

```python
m.save('tmb_map.html')
```
